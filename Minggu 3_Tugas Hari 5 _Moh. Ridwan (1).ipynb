````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here
# turn off spine/ticks
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')

ax = sns.scatterplot(x='carat', y='price', data=df, hue= 'clarity', palette=palette, size='depth', sizes=(1,6), linewidth=0,
                    hue_order=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'])
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style('dark'):
    facet_grid = sns.FacetGrid(df, hue='cut', col='cut', height=4)
    facet_grid = facet_grid.map(hexbin, 'carat', 'price', extent=[0, 3, 20000, 0.5])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
sns.pairplot(df, hue='cut', corner=True, hue_order=['Premium', 'Ideal', 'Very Good', 'Good', 'Fair'],
            markers=['+', 'o', 's', 'D', 'x'], vars=['carat', 'depth', 'table', 'price']);
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
import scipy.stats as stats
# Output 1
df = df.sample(n=300, random_state=123)
with sns.axes_style('white'):
    sns.jointplot(df['carat'], df['price'], kind='kde')
# Output 2
sns.jointplot("carat", "price", data=df, kind='reg', stat_func=stats.pearsonr, xlim=(0,2.6), ylim=(0,21000));
````
