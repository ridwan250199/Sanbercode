```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 8), sharey=True)
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, sin, color='r', label='sinX')
ax[2].plot(x, cos, color='g', label='cosX')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(15,9))
ins1 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
ins2 = ax.inset_axes([0.9,-0.1,0.3,0.3])
ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4 = ax.inset_axes([-0.1,0.9,0.3,0.3])
ax.plot(x, y, color='cornflowerblue')
ins1.plot(x, y,color='b')
ins2.plot(x, y,color='b')
ins3.plot(x, y,color='b')
ins4.plot(x, y,color='b')
ins1.plot(x, y2,color='r')
ins2.plot(x, y3,color='g')
ins3.plot(x, y4,color='k')
ins4.plot(x, y5,color='y')
ax.set_title('Grafik CosX')
ins1.set_title('Grafik CosX^2')
ins2.set_title('Grafik CosX^3')
ins3.set_title('Grafik CosX^4')
ins4.set_title('Grafik CosX^5')
plt.show()
```
