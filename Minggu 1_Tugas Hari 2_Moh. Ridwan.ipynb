<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

1. built-in Function merupakan fungsi yang sudah disediakan di Python untuk digunakan dengan kegunaan tertentu
2. sum(), len(), dan  max()




```python
# Contoh penggunaan built in function
a = [1,2,3,4]

#penggunaan sum() untuk menjumlahkan data/entery dalam list "a"
print(sum(a))

#penggunaan len() untuk menentukan panjang list atau banyaknya data dalam list "a"
print(len(a))

#penggunaan max() untuk menentukan nilai maksimum atau terbesar dari list "a"
print(max(a))
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

1. Method dipanggil melalui object. Karena method dipanggil pada object, ia dapat mengakses data di dalamnya sehingga method memungkinkan dapat mengubah status object. 

2. Function dipanggil tanpa melalui object sehingga function tidak dapat mengubah status object.



```python
#contoh method
class status(object):
   def my_method(self):
      print("Saya adalah seorang sarjana")
pribadi = status()
pribadi.my_method()
#dalam contoh tersebut, pribadi merupakan object untuk memanggil my_method()

#contoh function
def my_function(nama="Ridwan"):
      print(f"{nama} adalah seorang fresh graduate")
my_function()
#pemanggilan my_function() tanpa menggunakan object
```

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]
def kali_list(obj_list, obj_penambah):
    # isikan kode
    return obj_list + obj_penambah
    
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
