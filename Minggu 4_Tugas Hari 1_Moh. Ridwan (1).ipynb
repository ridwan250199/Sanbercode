#1
- Machine learning merupakan salah satu cabang dari ilmu kecerdasan buatan, khususnya yang mempelajari bagaimana komputer mamou belajar dari data untuk meningkatkan kecerdasannya
- Machine learning adalah komputer yang memiliki kemampuan melakukan belajar dari pengalaman terhadap tugas tugasnya di masa lalu dan mengalami peningkatan kerja

#2
- data feature (predictor) merupakan data yang memprediksi suatu kejadian
- data target (label) merupakan data hasil prediksi

#3
Supervised Learning
- Algoritma machine learning dengan menggunakan data latih untuk melakukan pembelajaran
- Data latihnya adalah data yang sudah berlabel
- Bertujuan agar mesin mampu mengidentifikasi label input yang baru dengan menggunakan fitur yang ada untuk melakuan prediksi maupun klaifikasi
- Digunakan untuk prediktif model
- Jenis algoritma : decision tree, random forest, regresi, naive bayes classifier, nearest neighboor, support vector machine, artificial neural network
Unsupervised Learning
- Algoritma machine learning yang tidak menggunakan data latih dalam melakukan oembelajaran
- Sistem mempelajari unlabeled data berdasarkan fitur fitur dari data tersebut
- Algoritma ini tidak memiliki variabel target dan bertujuan mengelompokkan objek yang serupa dalam suatu area tertentu
- Digunakan untuk clustering model
- Jenis algoritma : clustering, K-means, DBSCAN, fuzzy C-Means, self-organizing map

#4 Jenis Supervised Learning:
    
- Klasifikasi (outputnya: class label)
- Regresi (outputnya: kuantitas yang kontinu)

#5 Perbedaan Hyperparameter dan Parameter
- Hyperparameter merupakan bagian dari parameter. jadi, hyperparameter sudah pasti parameter, didefinisikan diawal pembuatan object model, tidak ada sangkut pautnya dengan model(tidak menempel dengan model, sesuatu yang didefinisikan untuk mengoptimasikan model) contoh: fit_intercept, dsb
- Parameter belum tentu hyperparameter, dihasilkan diakhir, ada sangkut pautnya dengan model(sesuatu yang berada dalam model), contoh: coef_ dan intercept_

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# Code here
#1
from sklearn.tree import DecisionTreeRegressor
```

```python
#2
model_dtr = DecisionTreeRegressor(max_depth=4, min_samples_leaf= 0.1)
```

```python
#3
x_matriks = x[:, np.newaxis]
```

```python
#4
model_dtr.fit(x_matriks, y)
```

```python
#5
x_test = np.linspace(20, 30, 15)
x_test = x_test[:, np.newaxis]
```

```python
y_test = model_dtr.predict(x_test)
```

```python
y_train = model_dtr.predict(x_matriks)
```

```python
fig, ax= plt.subplots(figsize=(16,10))
plt.scatter(x, y, label='Data Asli', color='b')
plt.scatter(x, y_train, label='Model Training', color='y')
plt.plot(x_test, y_test, label='Test Result/Hasil Prediksi', color='g')
plt.title('Algoritma DecisionTreeRegressor')
plt.legend()
plt.show()
```

```python
from sklearn.tree import plot_tree
fig, ax = plt.subplots(figsize=(12,8))
plot_tree(model_dtr)
plt.show()
```
